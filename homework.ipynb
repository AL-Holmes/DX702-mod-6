```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
from sklearn.neighbors import NearestNeighbors
```

```python
hw1 = pd.read_csv('homework_1.1.csv')
hw1
```

# Question 1

```python
X_hw1 = sm.add_constant(hw1.drop('Y',axis=1))
model = sm.OLS(hw1['Y'], X_hw1).fit()
print("--- Regression ---")
print(model.summary())
```

# Question 2

```python
differences = {}

for xi in hw1.drop('Y',axis=1).columns:
    # Marginal: Y ~ Xi (not fixing others)
    X_single = sm.add_constant(hw1[xi])
    coeff_marginal = sm.OLS(hw1['Y'], X_single).fit().params[xi]
    
    # Partial: Y ~ X1 + X2 + X3 (fixing others)
    coeff_partial = model.params[xi]
    
    differences[xi] = abs(coeff_marginal - coeff_partial)

# Find the Xi with the greatest difference
max_xi = max(differences, key=differences.get)
print('---max---')
print(max_xi)
print(f"The variable with the greatest difference is: {max_xi}")
```

```python
hw2 = pd.read_csv('homework_1.2.csv')
X_0 = hw2[hw2['X'] == 0].copy().reset_index(drop=True)
X_1 = hw2[hw2['X'] == 1].copy().reset_index(drop=True)
```

```python
nn = NearestNeighbors(n_neighbors=1)
nn.fit(X_0[['Z']])
dist, indexs = nn.kneighbors(X_1[['Z']])


print("---distances---")
print(len(dist))
print("---index---")
print(len(indexs))

match_hw2 = X_0.iloc[indexs.flatten()].copy().reset_index(drop=True)
y_1_mean = X_1['Y'].mean()
control_mean = match_hw2['Y'].mean()
effect_mean = y_1_mean - control_mean

print("---Matches---")
print(match_hw2)
print("---max distance---")
print(max(dist))
print()
print("---Effect---")
print(effect_mean)
```

```python
Z_0 = X_0[['Z']].values
Z_1 = X_1[['Z']].values
nnr = NearestNeighbors(radius=0.2).fit(Z_0)
dist_z, inx = nnr.radius_neighbors(Z_1)
print(f'Lenght of distances: {len(dist_z)}')
print(dist_z)
print()
print(len(inx))

control_y = np.average(X_0['Y'])
mean_y = X_1['Y'].mean()

print(f'Mean: {mean_y - control_y}')
```

```python
matches = []
for i, neighbor_indices in enumerate(inx):
    for j, neighbor_idx in enumerate(neighbor_indices):
        origin_idx = X_1.index[i]
        matched_idx = X_0.index[neighbor_idx]
        
        matches.append({
            'X=1_Index': origin_idx,
            'X=1_Z_Value': Z_1[i][0],
            'Matched_X=0_Index': matched_idx,
            'Matched_X=0_Z_Value': Z_0[neighbor_idx][0],
            'Distance': dist_z[i][j]
        })

results_df = pd.DataFrame(matches)
print(results_df)
```
